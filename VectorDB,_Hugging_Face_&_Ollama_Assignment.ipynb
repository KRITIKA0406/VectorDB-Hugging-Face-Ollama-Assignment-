#1.What is a Vector Database (VectorDB) and how is it different from
traditional databases?
Ans-A Vector Database (VectorDB) is a specialized database designed to store, index, and search vector embeddings — numerical representations of data such as text, images, audio, or videos generated by machine learning models.

These vectors capture the semantic meaning of data. For example, two sentences with similar meanings will have vectors that are mathematically close to each other.

What is a Vector?

In AI systems, data is converted into arrays of numbers called vectors (embeddings).

Example:

Sentence: “Artificial Intelligence is powerful”
Embedding Vector:
[0.12,−0.45,0.89,0.33,…]

These vectors may contain hundreds or thousands of dimensions.

Purpose of a Vector Database

A VectorDB is optimized for:

Storing embeddings efficiently
Performing similarity search
Retrieving semantically related data quickly
Supporting AI applications like:
Chatbots
Recommendation systems
Semantic search
Image retrieval
Retrieval-Augmented Generation (RAG)

Popular VectorDB systems include:

Pinecone
Weaviate
Milvus
Chroma
Qdrant
How VectorDB Works
Step 1: Data Conversion into Embeddings

An AI embedding model converts data into vectors.

Example:

Text → text embeddings
Images → image embeddings
Step 2: Store Vectors

The vectors are stored along with metadata.

Example:

ID	Text	Vector
1	“AI is transforming healthcare”	[0.23, 0.88, …]
2	“Machine learning improves diagnosis”	[0.21, 0.84, …]
Step 3: Similarity Search

When a user submits a query, it is also converted into a vector.

The database then finds vectors that are closest using similarity metrics such as:

Cosine similarity
Euclidean distance
Dot product
Difference Between Vector Databases and Traditional Databases
Feature	Vector Database	Traditional Database
Data Type	Vector embeddings	Structured rows and columns
Search Type	Semantic similarity search	Exact keyword/value matching
Main Purpose	AI/ML applications	Transactional and business applications
Query Style	“Find similar items”	SQL-based exact queries
Indexing Method	ANN (Approximate Nearest Neighbor) indexes	B-tree, hash indexes
Handles Unstructured Data	Very well	Limited
Example Use	Semantic search	Banking systems, ERP, CRM
Distance Metrics	Cosine, Euclidean	Not applicable
Example Comparison
Traditional Database Search

Query:

“Find documents containing the word ‘car’”

Results:

Only documents with the exact keyword “car”
Vector Database Search

Query:

“Find documents about automobiles”

Results may include:

“Cars”

Even if the exact word is absent, semantic meaning is understood.

Core Technologies Used in VectorDBs

Many VectorDBs use efficient search algorithms such as:

HNSW (Hierarchical Navigable Small World)
FAISS
IVF (Inverted File Index)
Approximate Nearest Neighbor (ANN)

These allow fast similarity search in very high-dimensional spaces.

Applications of Vector Databases
1. Semantic Search

Search based on meaning rather than exact words.

Example:

Search engines
Enterprise document retrieval
2. Recommendation Systems

Used by platforms like:

Netflix
Amazon
Spotify

to recommend similar products, movies, or songs.

3. Retrieval-Augmented Generation (RAG)

LLMs retrieve relevant documents from a VectorDB before generating responses.

Widely used in AI assistants and enterprise chatbots.

4. Image and Video Search

Search using image similarity instead of filenames.

5. Fraud Detection & Anomaly Detection

Find unusual patterns in data using vector similarity.

### Demonstrate Model Download and Fine-tuning for Text Generation

This section will show how to download a pre-trained Hugging Face model and fine-tune it on a small custom dataset for a text generation task.

```python
# Install necessary libraries if not already installed
!pip install transformers datasets accelerate peft bitsandbytes

import torch
from transformers import AutoModelForCausalLM, AutoTokenizer, TrainingArguments, Trainer
from datasets import Dataset
```

#### 1. Load Pre-trained Model and Tokenizer

We'll use a small `gpt2` model for demonstration purposes due to resource constraints. For larger tasks, you would choose a more powerful model.

```python
model_name = "gpt2"
tokenizer = AutoTokenizer.from_pretrained(model_name)
model = AutoModelForCausalLM.from_pretrained(model_name)

# Set the pad_token for the tokenizer if it's not already set (common for GPT-like models)
if tokenizer.pad_token is None:
    tokenizer.pad_token = tokenizer.eos_token
    model.config.pad_token_id = tokenizer.eos_token_id

print(f"Model '{model_name}' and tokenizer loaded successfully.")
```

#### 2. Prepare a Custom Dataset for Fine-tuning

We'll create a very small synthetic dataset for demonstration. In a real-world scenario, you would use a more extensive and relevant dataset.

```python
custom_data = [
    "The quick brown fox jumps over the lazy dog. This is a classic pangram.",
    "Artificial Intelligence is rapidly advancing and changing many industries.",
    "Hugging Face provides excellent tools for working with state-of-the-art NLP models.",
    "Fine-tuning a model adapts it to specific tasks and datasets.",
    "Generative AI can create new content, such as text, images, and music."
]

# Convert list of strings to a Dataset object
dataset = Dataset.from_dict({"text": custom_data})

print("Custom dataset created:")
print(dataset)
```

#### 3. Tokenize the Dataset

Before training, the text data needs to be converted into numerical tokens that the model can understand.

```python
def tokenize_function(examples):
    return tokenizer(examples["text"], truncation=True, padding="max_length", max_length=128)

tokenized_dataset = dataset.map(tokenize_function, batched=True)

# Remove the original 'text' column and rename 'input_ids' to 'labels' for causal language modeling
tokenized_dataset = tokenized_dataset.remove_columns(["text"])
tokenized_dataset = tokenized_dataset.rename_column("input_ids", "labels")

print("Tokenized dataset:")
print(tokenized_dataset)
```

#### 4. Set up Training Arguments and Trainer

Configure the training process, including output directory, learning rate, and epochs.

```python
training_args = TrainingArguments(
    output_dir="./results",
    learning_rate=2e-5,
    per_device_train_batch_size=2,  # Reduce batch size if you face OOM errors
    num_train_epochs=3,
    weight_decay=0.01,
    logging_dir="./logs",
    logging_steps=10,
    push_to_hub=False,
    save_strategy="no", # Don't save checkpoints to save space
)

trainer = Trainer(
    model=model,
    args=training_args,
    train_dataset=tokenized_dataset,
)

print("Trainer initialized.")
```

#### 5. Fine-tune the Model

Start the fine-tuning process. This will take a few moments.

```python
trainer.train()

print("Model fine-tuning complete!")
```

#### 6. Generate Text with the Fine-tuned Model

Now, let's use the fine-tuned model to generate some text based on a prompt. We expect it to be biased towards the topics in our small training dataset.

```python
prompt = "Artificial Intelligence is"

# Create a pipeline for text generation with the fine-tuned model
generator = pipeline("text-generation", model=model, tokenizer=tokenizer)

# Generate text
fine_tuned_result = generator(
prompt,
    max_new_tokens=50,  # Generate up to 50 new tokens
    num_return_sequences=1,
    truncation=True,
    do_sample=True,      # Enable sampling for more diverse output
    temperature=0.7,     # Control randomness
    top_k=50,            # Filter by top-k tokens
    top_p=0.95,          # Filter by tokens whose cumulative probability exceeds p
)

print(f"\nPrompt: {prompt}")
print("\nGenerated text (fine-tuned model):")
print(fine_tuned_result[0]['generated_text'])
```

```python
# Import libraries
import chromadb
from sentence_transformers import SentenceTransformer

# -----------------------------
# Step 1: Create Chroma Client
# -----------------------------
client = chromadb.Client()

# -----------------------------
# Step 2: Create Collection
# -----------------------------
collection = client.create_collection(name="semantic_search_demo")

# -----------------------------
# Step 3: Load Embedding Model
# -----------------------------
model = SentenceTransformer('all-MiniLM-L6-v2')

# -----------------------------
# Step 4: Sample Documents
# -----------------------------
documents = [
    "Artificial Intelligence is transforming industries.",
    "Machine learning enables computers to learn from data.",
    "Chroma DB is used for vector storage and semantic search.",
    "Natural Language Processing helps computers understand text.",
    "Deep learning models require large amounts of data."
]

# Unique IDs for documents
ids = ["doc1", "doc2", "doc3", "doc4", "doc5"]

# -----------------------------
# Step 5: Generate Embeddings
# -----------------------------
embeddings = model.encode(documents).tolist()

# -----------------------------
# Step 6: Insert Data into Chroma DB
# -----------------------------
collection.add(
    documents=documents,
    embeddings=embeddings,
    ids=ids
)

print("Documents inserted successfully into Chroma DB!")

# -----------------------------
# Step 7: Perform Semantic Search
# -----------------------------
query = "What database is useful for semantic search?"

# Generate query embedding
query_embedding = model.encode([query]).tolist()

# Search similar documents
results = collection.query(
    query_embeddings=query_embedding,
    n_results=2
)

# -----------------------------
# Step 8: Display Results
# -----------------------------
print("\nQuery:", query)

print("\nTop Matching Documents:")
for i, doc in enumerate(results['documents'][0]):
    print(f"{i+1}. {doc}")
```

This example demonstrates how to:

Prepare a small custom dataset

Generate text using the fine-tuned model

Python Code

```python
# ---------------------------------------
# Step 1: Import Required Libraries
# ---------------------------------------
from datasets import Dataset
from transformers import (
    AutoTokenizer,
    AutoModelForCausalLM,
    Trainer,
    TrainingArguments,
    DataCollatorForLanguageModeling
)

# ---------------------------------------
# Step 2: Sample Training Data
# ---------------------------------------
texts = [
    "Artificial Intelligence is changing the world.",
    "Machine learning helps computers learn from data.",
    "Natural Language Processing enables text understanding.",
    "Deep learning is a subset of machine learning.",
    "Hugging Face provides powerful NLP models."
]

# Convert into Hugging Face Dataset
dataset = Dataset.from_dict({"text": texts})

# ---------------------------------------
# Step 3: Load Pre-trained Model
# ---------------------------------------
model_name = "distilgpt2"

tokenizer = AutoTokenizer.from_pretrained(model_name)
model = AutoModelForCausalLM.from_pretrained(model_name)

# GPT-2 models do not have pad token by default
tokenizer.pad_token = tokenizer.eos_token

# ---------------------------------------
# Step 4: Tokenization Function
# ---------------------------------------
def tokenize_function(example):
    return tokenizer(
        example["text"],
        padding="max_length",
        truncation=True,
        max_length=50
    )

tokenized_dataset = dataset.map(tokenize_function)

# ---------------------------------------
# Step 5: Data Collator
# ---------------------------------------
data_collator = DataCollatorForLanguageModeling(
    tokenizer=tokenizer,
    mlm=False
)

# ---------------------------------------
# Step 6: Training Arguments
# ---------------------------------------
training_args = TrainingArguments(
    output_dir="./fine_tuned_model",
    overwrite_output_dir=True,
    num_train_epochs=3,
    per_device_train_batch_size=2,
    save_steps=10,
    save_total_limit=1,
    logging_steps=5
)

# ---------------------------------------
# Step 7: Trainer Setup
# ---------------------------------------
trainer = Trainer(
    model=model,
    args=training_args,
    train_dataset=tokenized_dataset,
    data_collator=data_collator
)

# ---------------------------------------
# Step 8: Fine-Tune the Model
# ---------------------------------------
trainer.train()

# ---------------------------------------
# Step 9: Save Fine-Tuned Model
# ---------------------------------------
trainer.save_model("./fine_tuned_model")

print("Model fine-tuning completed!")

# ---------------------------------------
# Step 10: Generate Text
# ---------------------------------------
input_text = "Artificial Intelligence"

inputs = tokenizer(input_text, return_tensors="pt")

output = model.generate(
    **inputs,
    max_length=40,
    num_return_sequences=1
)

generated_text = tokenizer.decode(output[0], skip_special_tokens=True)

print("\nGenerated Text:")
print(generated_text)
```

### RAG Implementation

Now, let's implement the RAG system. This code will:

1.  Define a small knowledge base (sample documents).
2.  Use `SentenceTransformer` to create embeddings for these documents.
3.  Store these embeddings and documents in Chroma DB.
4.  Define a `rag_ollama` function that takes a query, retrieves relevant documents from Chroma DB, and then uses the `ollama` Python client to query Llama3 (assuming an Ollama server is running with Llama3 pulled).
5.  Demonstrate the RAG system with a sample query.

```python
import ollama
import chromadb
from sentence_transformers import SentenceTransformer

# --- Step 1: Define a Knowledge Base (Sample Documents) ---
knowledge_base_documents = [
    "Large Language Models (LLMs) are deep learning models trained on vast amounts of text data.",
    "Retrieval-Augmented Generation (RAG) combines information retrieval with text generation.",
    "Chroma DB is an open-source vector database designed for AI applications.",
    "Ollama allows users to run open-source large language models locally.",
    "Llama3 is a powerful open-source large language model developed by Meta.",
    "Vector databases store numerical representations of data, enabling semantic search.",
    "Semantic search allows finding information based on meaning rather than keywords.",
    "Fine-tuning adapts a pre-trained model to a specific task or dataset."
]

# --- Step 2 & 3: Create Embeddings and Store in Chroma DB ---
# Initialize SentenceTransformer for embeddings
embedding_model = SentenceTransformer('all-MiniLM-L6-v2')

# Create a Chroma client (in-memory for this example)
chroma_client = chromadb.Client()

# Create a collection
collection_name = "llama3_rag_knowledge"
try:
    chroma_collection = chroma_client.create_collection(name=collection_name)
except: # Collection might already exist if cell is re-run
    chroma_collection = chroma_client.get_collection(name=collection_name)

# Generate embeddings and prepare documents for Chroma DB
document_embeddings = embedding_model.encode(knowledge_base_documents).tolist()
document_ids = [f"doc_{i}" for i in range(len(knowledge_base_documents))]

# Add documents to Chroma DB
chroma_collection.add(
    documents=knowledge_base_documents,
    embeddings=document_embeddings,
    ids=document_ids
)

print(f"Knowledge base loaded into Chroma DB collection: {collection_name}")

# --- Step 4: Define the RAG function ---
def rag_ollama(query, ollama_model="llama3", n_results=2):
    # Generate embedding for the query
    query_embedding = embedding_model.encode([query]).tolist()

    # Retrieve relevant documents from Chroma DB
    retrieved_results = chroma_collection.query(
        query_embeddings=query_embedding,
        n_results=n_results,
        include=['documents']
    )

    # Extract relevant documents
    relevant_documents = retrieved_results['documents'][0]

    # Construct the prompt for Llama3
    context = "\n".join(relevant_documents)
    prompt = f"Based on the following context, answer the question:\n\nContext:\n{context}\n\nQuestion: {query}\n\nAnswer:"

    try:
        # Use ollama to get a response from Llama3
        # Ensure Ollama server is running and llama3 model is pulled
        response = ollama.chat(
            model=ollama_model,
            messages=[
                {'role': 'user', 'content': prompt}
            ]
        )
        return response['message']['content']
    except ollama.ConnectionError:
        return "Error: Could not connect to Ollama server. Please ensure Ollama is running and Llama3 is pulled."
    except Exception as e:
        return f"An error occurred during Ollama interaction: {e}"

print("RAG function defined.")

# --- Step 5: Demonstrate RAG ---
print("\n--- Demonstrating RAG with Llama3 ---")
query_rag = "What is RAG and how does it relate to LLMs?"
print(f"\nQuery: {query_rag}")

rag_response = rag_ollama(query_rag)
print("\nGenerated Response:")
print(rag_response)

print("\n--- Another RAG Query ---")
query_ollama = "What is Ollama used for?"
print(f"\nQuery: {query_ollama}")

ollama_response = rag_ollama(query_ollama)
print("\nGenerated Response:")
print(ollama_response)
```

```python
# ---------------------------------------------------
# Step 1: Import Libraries
# ---------------------------------------------------
import chromadb
from sentence_transformers import SentenceTransformer
import ollama

# ---------------------------------------------------
# Step 2: Initialize ChromaDB
# ---------------------------------------------------
client = chromadb.Client()

collection = client.create_collection(
    name="medical_articles"
)

# ---------------------------------------------------
# Step 3: Load Hugging Face Embedding Model
# ---------------------------------------------------
embedding_model = SentenceTransformer(
    'all-MiniLM-L6-v2'
)

# ---------------------------------------------------
# Step 4: Sample Medical Research Articles
# ---------------------------------------------------
documents = [
    "Diabetes can be managed through insulin therapy and healthy diet.",

    "COVID-19 vaccines help reduce severe infections and hospitalization.",

    "Regular exercise improves cardiovascular health and reduces stress.",

    "Early detection of cancer significantly increases treatment success rates."
]

ids = ["doc1", "doc2", "doc3", "doc4"]

# ---------------------------------------------------
# Step 5: Generate Embeddings
# ---------------------------------------------------
embeddings = embedding_model.encode(
    documents
).tolist()

# ---------------------------------------------------
# Step 6: Store in ChromaDB
# ---------------------------------------------------
collection.add(
    documents=documents,
    embeddings=embeddings,
    ids=ids
)

print("Medical articles stored successfully!")

# ---------------------------------------------------
# Step 7: User Query
# ---------------------------------------------------
query = "How do vaccines help against COVID?"

# Generate query embedding
query_embedding = embedding_model.encode(
    [query]
).tolist()

# ---------------------------------------------------
# Step 8: Retrieve Relevant Articles
# ---------------------------------------------------
results = collection.query(
    query_embeddings=query_embedding,
    n_results=2
)

retrieved_docs = results['documents'][0]

print("\nRetrieved Articles:")
for doc in retrieved_docs:
    print("-", doc)

# ---------------------------------------------------
# Step 9: Build Context for LLM
# ---------------------------------------------------
context = "\n".join(retrieved_docs)

prompt = f"""
You are a medical assistant chatbot.

Use the following medical research information to answer the question.

Research Information:
{context}

Question:
{query}

Answer:
"""

# ---------------------------------------------------
# Step 10: Generate Response Using Ollama
# ---------------------------------------------------
response = ollama.chat(
    model='llama2',
    messages=[
        {
            'role': 'user',
            'content': prompt
        }
    ]
)

# ---------------------------------------------------
# Step 11: Display Final Answer
# ---------------------------------------------------
print("\nChatbot Response:")
print(response['message']['content'])
```
